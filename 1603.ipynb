```python
import pandas as pd
```

```python
data = pd.read_excel('1503test.xlsx', engine='openpyxl')
```

```python
data.head()
```

```python
import json
# Tạo danh sách các dictionary chứa thông tin về counsel
result = []

for index, row in data.iterrows():

    counsel = []
    if not pd.isna(row["counsel_for_the_applicant"]):
        counsel.append({
            "name": row["counsel_for_the_applicant"],
            "role": "for the applicant",
            "else": ""
        })
    else:
        counsel.append({
            "name": "",
            "role": "",
            "else": ""
        })

    if not pd.isna(row["counsel_for_the_respondent"]):
        counsel.append({
            "name": row["counsel_for_the_respondent"],
            "role": "for the respondent",
            "else": ""
        })
    else:
        counsel.append({
            "name": "",
            "role": "",
            "else": ""
        })

    soli = []
    if not pd.isna(row["solicitors_applicant"]):
        soli.append({
            "name": row["solicitors_applicant"],
            "role": "for the applicant",
            "else": ""
        })
    else:
        soli.append({
            "name": "",
            "role": "",
            "else": ""
        })

    if not pd.isna(row["solicitors_respondent"]):
        soli.append({
            "name": row["solicitors_respondent"],
            "role": "for the respondent",
            "else": ""
        })
    else:
        soli.append({
            "name": "",
            "role": "",
            "else": ""
        })

    ju = []
    if not pd.isna(row["federal_court"]):
        ju.append(
            {
	            "location": '',
	            "name": [row['name']],
	            "organization": [row['federal_court']]
            }
        )
    else:
        ju.append({
            "location": "",
            "name": [row['name']],
            "organization": ""
        })

    record = {
        'id' : row['id'],
        'counsel': counsel,
        'solicitors':soli,
        'judge' : ju
    }

    result.append(record)


json_data = json.dumps(result, indent=4)

# Lưu dữ liệu vào một file JSON
with open('output.json', 'w', encoding='utf-8') as f:
    f.write(json_data)

    
```

```python

import os

# Đọc dữ liệu từ file JSON
with open('output.json', 'r') as json_file:
    data = json.load(json_file)

# Đường dẫn đầy đủ đến thư mục lưu trữ các file txt
output_folder = 'test'

# Kiểm tra và tạo thư mục nếu nó chưa tồn tại
os.makedirs(output_folder, exist_ok=True)

# Lưu mỗi bản ghi thành một file .txt
for record in data:
    filename = os.path.join(output_folder, f"{record['id']}")
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(json.dumps(record, ensure_ascii=False, indent=4))
```
